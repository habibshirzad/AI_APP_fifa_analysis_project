### Query Four
Now that they have some information about the players they want us to go deeper and findout a list of the players who has German citizenship, not older than 30 and their salary is over 350k.

to do that we use comparison operators or conditional statements use the required column names such as wage_euro, nationality and age, and then assign it to a variable and then print its value.

```python
players_under_30 = df[(df['age'] < 30) & (df['wage_euro'] > 350000) & (df['nationality'] == 'Germany') ]
players_under_30.head(5)
```

### Query Five
Union berlin football club wants to hire a player whos under age 20, and his dribling skill is greater than 75 and 
speed is greater than 80.
to print out the list of the players with such skills we use conditional statement and logical operators to combine three or more skills and then assign it to a variable and then print it. and as result we can see that there's only one player with the required skills.

```python
FCB_striker_or_mid = df[(df['age']<20) & (df['dribbling']>75) & (df['sprint_speed']>80)]
FCB_striker_or_mid
```

### Query Six
Berlin Footbal Club is looking for a list of the players with the following; experienced MIDFIELDER who can play as LM,RM','CM','ST OR 'CAM','LW','RW'. also their age should be higher than 20 and salary greater 150k.

we use logical operators and isin method to search for the players can play in the specific positions.

after running the query we can see that there are only two players that matches all the required skills and conditions.

```python
FCB_MID = df[(df['positions'].isin(['LM,RM','CM','ST']) | df['positions'].isin(['CAM','LW','RW'])) & (df['wage_euro'] > 150000) & (df['age'] > 20) & (df['dribbling'] > 75) & (df['sprint_speed'] > 80 ) ]
FCB_MID
```

### Query Seven
FCB is looking for a defender who's tackle is greater than 75, jumping greater than 80, and marking is also greater than 80.

as we can see below in the table the details of the players that matches all the requirements are assigned to a variable called FCB_DEF.

```python
FCB_DEF = df[(df['standing_tackle']>75) & (df['heading_accuracy']>75) & (df['jumping']>80) & (df['marking']>80)]
FCB_DEF
```

### Query 8
A list of the top 3 attacking players with their name, value and jersey number is required.

To do that, this time we are using LAMBDA function to write our query. 
first we sort our data set using sort method to order our dataset based on the top players which is overall_rating. and then assign it to a variable called top_3.  and then apply the lambda dunction on this variable,access teh calumns and their values.


```python
top_3 = df.sort_values(by='overall_rating', ascending=False)

nationality = top_3.apply(lambda x: (x['national_jersey_number'],x['name'] ,x['nationality'],x['value_euro']), axis=1)
nationality.head(3)
```

### Query 9
FCB is looking for a reserved attacking player who can play in any attacking positions and wants to make sure that they can afford them. therefore they need to know their name and salary.

to do that first we create a variable and then filter out and apply all the conditions and then we use variable to print the name and their salaries and we also apply head method to print the first 5. 

```python
Attacking_Player = df[(df['positions'].isin(['LM,RM,RW,ST,CF'])) & (df['age'] > 20) & (df['dribbling'] > 75) | (df['sprint_speed'] > 80)]
Attacking_Player[['name','wage_euro']].head()
```

### Query 10

printing a list of the Players whos salary is in between 150k and 250k and plays LM', 'RM', 'RW','ST'.

to print the list we need isin method, conditional and logical statements. below we access all the details for example if we want to name of these players all we need to do is salary['name']. it will print the names of the players that we filtered.

```python
salary = df[(df['wage_euro'] < 250000) & (df['wage_euro'] > 100000) & df['positions'].isin(['LM', 'RM', 'RW', 'ST'])]
salary
```
